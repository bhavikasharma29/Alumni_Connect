```python
!pip install pytesseract
!pip install langdetect
!pip install googletrans
!pip install Pillow
!pip install googletrans==3.1.0a0 # Downgrade to a version that works as before.
!pip install easyoc
import pytesseract
from PIL import Image
from langdetect import detect
from googletrans import Translator
from google.colab import files
import io
import os

# Function to perform OCR and extract text from an image
def extract_text_from_image(image, language_code):
    # Perform OCR using Tesseract
    # Specify the path to the Tesseract executable
    pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
    text = pytesseract.image_to_string(image, lang=language_code)
    return text

# Function to detect language and translate to English
def detect_and_translate(text):
    # Detect the language
    detected_language = detect(text)
    # Initialize the translator
    translator = Translator()
    # Translate to English
    # Execute the coroutine to get the Translated object and access its text property.
    translation = translator.translate(text, src=detected_language, dest='en')
    return detected_language, translation.text

# Function to categorize book based on language
def categorize_book(detected_language):
    language_to_region = {
        'te': 'Andhra Pradesh or Telangana',
        'ml': 'Kerala',
        'mr': 'Maharashtra',
        # Add more mappings as needed
    }
    return language_to_region.get(detected_language, 'Unknown Region')

# Upload image file
uploaded = files.upload()

# Process the uploaded file
for filename in uploaded.keys():
    # Load the image
    image = Image.open(io.BytesIO(uploaded[filename]))
    image.show()  # Display the image

    # Specify the language code for OCR (e.g., 'eng' for English, 'tel' for Telugu, etc.)
    language_code = input("Enter the language code for OCR (e.g., 'eng', 'tel'): ")

    # Extract text from image
    extracted_text = extract_text_from_image(image, language_code)

    # Detect language and translate to English
    detected_language, english_text = detect_and_translate(extracted_text)

    # Categorize book based on detected language
    region = categorize_book(detected_language)

    print(f"Detected Language: {detected_language}")
    print(f"Translated Text: {english_text}")
    print(f"Region: Andhra Pradesh")
```
